# Operational Stress and Anomaly Detection

## Objective

This notebook transitions from descriptive exploration to diagnostic analysis.

The objective is to identify **districts experiencing disproportionate operational stress** within the UIDAI system by examining imbalances between enrolment intake and update-driven maintenance activity.

### Key Framing Principles

- **Stress â‰  high volume**
- **Stress = imbalance relative to expectation**
- Comparisons are **relative**, not normative judgments

High activity alone does not imply stress. Instead, stress is defined as **update intensity that exceeds what would typically be expected for districts of similar scale**.

> This notebook focuses exclusively on **operational stress hotspots**.  
> Lifecycle access gaps and inclusion issues are analyzed separately in Notebook 4.

## Step 1: Load Cleaned District-Level Datasets

We begin by loading the cleaned and standardized district-level datasets produced in Notebook 1.

These datasets serve as the single source of truth for all stress calculations.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
enrol = pd.read_csv("../data/processed/enrolment_clean.csv", parse_dates=["date"])
demo  = pd.read_csv("../data/processed/demographic_updates_clean.csv", parse_dates=["date"])
bio   = pd.read_csv("../data/processed/biometric_updates_clean.csv", parse_dates=["date"])
```

## Step 2: Construct District-Level Master Table

To ensure comparability across datasets, all activity is aggregated across time to the **district level**.

This produces a unified master table containing:
- Total enrolments
- Total demographic updates
- Total biometric updates

This table is the analytical backbone for all subsequent stress calculations.

### Aggregating enrolments

```python
enrol_agg = enrol.groupby(["state", "district"], as_index=False)[
    ["age_0_5", "age_5_17", "age_18_greater"]
].sum()

enrol_agg["total_enrolments"] = (
    enrol_agg["age_0_5"] +
    enrol_agg["age_5_17"] +
    enrol_agg["age_18_greater"]
)
```

### Aggregating demographic updates

```python
demo_agg = demo.groupby(["state", "district"], as_index=False)[
    ["demo_age_5_17", "demo_age_17_plus"]
].sum()

demo_agg["total_demo_updates"] = (
    demo_agg["demo_age_5_17"] +
    demo_agg["demo_age_17_plus"]
)
```

### Aggregating biometric updates

```python
bio_agg = bio.groupby(["state", "district"], as_index=False)[
    ["bio_age_5_17", "bio_age_17_plus"]
].sum()

bio_agg["total_bio_updates"] = (
    bio_agg["bio_age_5_17"] +
    bio_agg["bio_age_17_plus"]
)
```

### Merging into master table

```python
master = (
    enrol_agg.merge(demo_agg, on=["state", "district"], how="inner")
             .merge(bio_agg,  on=["state", "district"], how="inner")
)

master = master[master["total_enrolments"] > 0]
master.head()
```

## Step 3: Update Intensity Ratios (UIR)

Raw volumes obscure stress because districts vary widely in size.

To normalize for scale, we compute **Update Intensity Ratios (UIR)**, defined as update activity relative to enrolment intake.

These ratios represent **maintenance burden per unit of enrolment**.

```python
master["demo_stress_ratio"] = (
    master["total_demo_updates"] / master["total_enrolments"]
)

master["bio_stress_ratio"] = (
    master["total_bio_updates"] / master["total_enrolments"]
)

master["total_update_ratio"] = (
    (master["total_demo_updates"] + master["total_bio_updates"]) /
    master["total_enrolments"]
)
```

### Age-Specific Update Intensity

Stress dynamics differ across lifecycle stages.

To capture this, we compute **age-specific update intensity ratios** for dominant lifecycle transitions:
- Adult demographic updates
- Adolescent and adult biometric updates

```python
master["demo_stress_17_plus"] = (
    master["demo_age_17_plus"] / master["age_18_greater"]
).replace([np.inf, -np.inf], np.nan)

master["bio_stress_5_17"] = (
    master["bio_age_5_17"] / master["age_5_17"]
).replace([np.inf, -np.inf], np.nan)

master["bio_stress_17_plus"] = (
    master["bio_age_17_plus"] / master["age_18_greater"]
).replace([np.inf, -np.inf], np.nan)
```

At this stage, we **do not rank districts**.

Instead, we inspect the distribution of stress ratios to identify extreme behavior and motivate a peer-based expectation framework.

```python
plt.figure()
plt.hist(master["demo_stress_ratio"], bins=50)
plt.title("Distribution of Demographic Update Intensity")
plt.xlabel("Demographic Update Intensity Ratio")
plt.ylabel("Number of Districts")
plt.show()
```

```python
# Compute global median demographic stress ratio
global_median = master["demo_stress_ratio"].median()

plt.figure()
plt.scatter(master["total_enrolments"], master["demo_stress_ratio"])
plt.axhline(global_median, linestyle="--", linewidth=2)

plt.xlabel("Total Enrolments")
plt.ylabel("Demographic Update Intensity Ratio")
plt.title("Enrolments vs Demographic Update Intensity")

plt.show()
```

The dashed line represents the national median demographic update intensity. 
Districts significantly above this baseline exhibit disproportionately high maintenance activity relative to enrolment intake.

## Step 4: Expected vs Observed Stress Framework

Districts of similar scale are expected to exhibit similar update behavior.

Stress is therefore defined as **deviation from peer expectations**, not deviation from national averages.

### Enrolment Peer Bins

```python
master["enrolment_bin"] = pd.qcut(
    master["total_enrolments"],
    q=3,
    labels=["Low Enrolment", "Medium Enrolment", "High Enrolment"]
)
```

### Expected Ratios (Peer Medians)

```python
# Expected (typical) demographic update intensity within each enrolment bin
master["expected_demo_ratio"] = (
    master.groupby("enrolment_bin", observed=False)["demo_stress_ratio"]
          .transform("median")
)

# Expected (typical) biometric update intensity within each enrolment bin
master["expected_bio_ratio"] = (
    master.groupby("enrolment_bin", observed=False)["bio_stress_ratio"]
          .transform("median")
)
```

### Deviations

```python
# Demographic stress deviation
master["demo_deviation"] = (
    master["demo_stress_ratio"] / master["expected_demo_ratio"]
)

# Biometric stress deviation (supporting signal)
master["bio_deviation"] = (
    master["bio_stress_ratio"] / master["expected_bio_ratio"]
)
```

### Visual Validation of Peer Expectations

The following visualization illustrates how demographic update intensity varies within enrolment peer groups, 
reinforcing the need for peer-relative stress assessment.

```python
plt.figure()
master.boxplot(
    column="demo_stress_ratio",
    by="enrolment_bin"
)

plt.title("Demographic Update Intensity by Enrolment Peer Group")
plt.suptitle("")
plt.xlabel("Enrolment Peer Group")
plt.ylabel("Demographic Update Intensity Ratio")
plt.show()
```

```python
master[[
    "state",
    "district",
    "enrolment_bin",
    "demo_stress_ratio",
    "expected_demo_ratio",
    "demo_deviation"
]].head(10)
```

## Step 5: Stress Scoring

Stress is formalized using **percentile-based thresholds**, avoiding arbitrary cutoffs.

Districts in the top decile of deviation are classified as high-stress areas.

```python
stress_threshold = master["demo_deviation"].quantile(0.90)

master["high_stress_flag"] = master["demo_deviation"] >= stress_threshold
```

## Step 6: Identifying Stress Hotspots

We now identify districts exhibiting the highest operational stress and assess the concentration of system load they represent.

```python
top_stress = (
    master.sort_values("demo_deviation", ascending=False)
          .head(10)
          .reset_index(drop=True)
)

top_stress[[
    "state",
    "district",
    "enrolment_bin",
    "demo_stress_ratio",
    "expected_demo_ratio",
    "demo_deviation",
    "total_update_ratio"
]]
```

### Interpreting Stress Hotspots

The table above lists districts exhibiting the highest **demographic stress deviations**, defined as update intensity far exceeding what is typical for districts of similar enrolment scale.

Key observations:

- Stress deviations range from **~5Ã— to over 20Ã— expected levels**, indicating extreme maintenance burden in select districts.
- These districts are not necessarily the largest by enrolment, reinforcing that **raw volume alone does not explain stress**.
- High total update ratios suggest that frontline UIDAI infrastructure in these areas is predominantly engaged in **maintenance and corrections rather than new enrolments**.

These districts represent **operational stress hotspots** where targeted staffing, infrastructure scaling, or temporary outreach interventions may yield outsized impact.

### Concentration Check

```python
stress_summary = pd.DataFrame({
    "Metric": [
        "High-stress districts (share of all districts)",
        "Demographic updates originating from high-stress districts"
    ],
    "Value": [
        f"{stress_district_share:.1%}",
        f"{stress_update_share:.1%}"
    ]
})

stress_summary.style.hide(axis="index")
```

### Concentration of Operational Stress

Although high-stress districts represent a limited share of all districts, they account for a **disproportionately large fraction of demographic update activity**.

This concentration indicates that UIDAIâ€™s operational workload is **not evenly distributed**, with localized pressure points that may not be apparent from national aggregates.

From an operational perspective, this suggests that **targeted interventions in a small number of districts could alleviate a meaningful portion of system-wide maintenance burden**, improving service efficiency without requiring uniform resource expansion.

```python
avg_updates_high_stress = (
    master.loc[master["high_stress_flag"], "total_demo_updates"].sum()
    / master["high_stress_flag"].sum()
)

avg_updates_other = (
    master.loc[~master["high_stress_flag"], "total_demo_updates"].sum()
    / (~master["high_stress_flag"]).sum()
)

plt.figure()
plt.bar(
    ["High-stress districts", "All other districts"],
    [avg_updates_high_stress, avg_updates_other]
)

plt.ylabel("Average Demographic Updates per District")
plt.title("Per-District Demographic Update Load")
plt.show()
```

**High-stress districts handle approximately 60% more demographic update activity per district compared to other districts**, highlighting localized operational pressure that may not be visible in aggregate statistics.

## Conclusion

Operational stress within the UIDAI system is **not driven by enrolment volume**, but by **concentrated demographic update activity**.

Key observations:

- Demographic update intensity is highly skewed across districts, making raw volumes an unreliable indicator of operational pressure.
- Districts of similar enrolment scale exhibit comparable â€œnormalâ€ maintenance behavior; stress emerges where update activity **deviates sharply from peer expectations**.
- A small subset of districts operates at several times the expected demographic update intensity and processes significantly higher update workloads per district.
- This concentration indicates **localized stress hotspots** where frontline infrastructure and staffing capacity are most likely to be strained.

The implication is clear: **uniform resource allocation obscures operational pressure**.  
Targeted intervention in identified stress hotspots offers a more efficient path to system stabilization.
